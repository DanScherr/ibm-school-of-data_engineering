### TODO: 4. Procure   na   internet   um   arquivo   .csv   e   transforme-o   em   DataFrame   utilizando   a biblioteca Pandas

```python
import pandas as pd
df = pd.read_csv("India_Menu.csv")
df.head()
```

```python
df.info()
```

```python
def remove_units (DataFrame, columns, args):
    for col in columns:
        for item in args:
            DataFrame[col] = DataFrame[col].str.strip(item)

remove_units(df, ['Per Serve Size'], [' ', 'g', 'ml'])
df.head()
```

```python
def to_type(DataFrame, columns, type):
    for col in columns:
        if type == "int":
            print("entrou")
            DataFrame[col] = DataFrame[col].round()
        DataFrame[col] = DataFrame[col].astype(type)

to_type(df, ['Per Serve Size'], 'float')
df.info()
```

        iii. Adicioná-la ao dataframe

```python
new_column = df['Per Serve Size'][:] * 1.0

def search_column_name(DataFrame, name):
    for index, header in enumerate(DataFrame.keys()):
        if header == name:
            return index
        
        
insert_index = search_column_name(df, 'Per Serve Size')
df.insert(insert_index, 'perServeSize', new_column)
df.head()
```
